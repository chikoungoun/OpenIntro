# Recreate the contingency table on the whole original "email" Dataset

```python
import pandas as pd
import numpy as np
import csv
```

### Treating the ".txt" file to convert it in ".csv"

```python
with open('Dataset/email.txt','r') as email:
    file = email.readlines()
    t=[]
    
    with open('Dataset/email.csv','w') as e:
        eWriter = csv.writer(e)
        
        for c,i in enumerate(file):
            t = i.split('\t')
            eWriter.writerow(t)
    e.close()
            
        
    
```

```python
df = pd.read_csv('Dataset/email.csv')
```

```python
df.head()
```

For some reasons all the last characters end up  with '\n' when getting put with 'csv methods'

```python
df = df.rename(columns = {'number\n':'number'})
```

```python
df.replace(inplace=True,to_replace=r'$\n',value='',regex=True)
```

###### Here goes the dataframe

```python
df.head()
```

```python
df.columns
```

```python
def counting(x):
    return x.count()
```

### Generate the contingency Table : 

```python
cont = df.pivot_table(index='spam',columns='format',values='num_char',aggfunc=[counting])
cont
```

with : 
- Format :
  * 0 : Text
  * 1 : HTML

- Spam : 
  * 0 : Not Spam
  * 1 : Spam

- Adding the total column

```python
cont.loc[:,('counting','total')] = cont.loc[:,('counting',0)] + cont.loc[:,('counting',1)]
cont
```

### Column Proportion

```python
cont.loc[:,('counting')].div(cont.loc[:,('counting','total')],axis=0)
```
