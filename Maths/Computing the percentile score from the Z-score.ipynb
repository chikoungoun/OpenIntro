# Computing the percentile score from the Z-score

```python
import math
import numpy as np
from scipy import integrate
```

We need to code this specific equation for the percentile score

**Cumulative Distribution Function**

```python
from IPython.display import Image
Image(filename="cdf.png")
```

We define the function in python code

```python
def fn(x):
    return (1/math.sqrt(2*math.pi)) * math.exp(-x**2/2)
```

Let's make some tests. We define the $-\infty$

```python
ninf = np.NINF
```

and make the test with $Z = 1.25$ : 

```python
z= 1.25
#the fucntion returns a tuple (percentile, approximate error)
percentile, err =integrate.quad(fn,ninf,z)
```

```python
round(percentile,4)
```
